```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/mohammadreza/Downloads/quera_crime/data/train.csv")
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')
df['DATE OCC'] = df['DATE OCC'].dt.month*100 + df['DATE OCC'].dt.day
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p')
df['Date Rptd'] = df['Date Rptd'].dt.month*100 + df['Date Rptd'].dt.day
df = df.drop(columns=['DR_NO','Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4', 
                      'Crm Cd Desc', 'Status Desc', 'Cross Street','Weapon Used Cd'
                      , 'Weapon Desc', 'LOCATION','Weapon Desc'], axis=1)

df.head(5)
```

```python
# Ordinal variables
df['Vict Sex'] = df['Vict Sex'].replace(['M', 'F', 'X', 'H'], [0, 1, 2, 3])
df['Status'] = df['Status'].replace(['AA', 'AO', 'JA', 'JO', 'IC'], [0, 1, 2, 3, 4])

# Dropping NA values and resetting the indexes each time
df['Premis Desc'] = df['Premis Desc'].dropna()
df = df.reset_index(drop=True)
df['Mocodes'] = df['Mocodes'].dropna()
df = df.reset_index(drop=True)
df['Vict Sex'] = df['Vict Sex'].dropna()
df = df.reset_index(drop=True)
df['Vict Descent'] = df['Vict Descent'].dropna()
df = df.reset_index(drop=True)
df['Premis Desc'] = df['Premis Desc'].dropna()
df = df.reset_index(drop=True)
# Transforming categorical values to numerical
le = LabelEncoder()
df['AREA NAME']= le.fit_transform(df['AREA NAME'])
df['Mocodes']= le.fit_transform(df['Mocodes'])
df['Vict Sex']= le.fit_transform(df['Vict Sex'])
df['Vict Descent']= le.fit_transform(df['Vict Descent'])
df['Premis Desc']= le.fit_transform(df['Premis Desc'])
df['Status']= le.fit_transform(df['Status'])
df['Crm Cd']= le.fit_transform(df['Crm Cd'])

df.info()
```

```python
from sklearn.experimental import enable_iterative_imputer 
from sklearn.impute import IterativeImputer, KNNImputer
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from xgboost import XGBRegressor
from sklearn.metrics import f1_score, accuracy_score, r2_score, classification_report
```

```python
KNNImputer(n_neighbors=5)
imp = IterativeImputer(estimator=DecisionTreeRegressor(), max_iter=10,random_state=0,n_nearest_features=5)
imp_df = imp.fit_transform(df)
imp_df = pd.DataFrame(imp_df,columns=df.columns)
```

```python
x1 = imp_df.iloc[:,:6]
x2 = imp_df.iloc[:,7:]
x = pd.concat([x1,x2],axis=1)
imp_df.iloc[:,6:]
y = pd.DataFrame(imp_df['Crm Cd'])

std = MinMaxScaler()
x = std.fit_transform(x)
y = std.fit_transform(y)

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.10, random_state=0, stratify=y)
```

```python
params = {"learning_rate" : [0.1,0.01,0.001,0.002,0.0001,0.0002],
          "n_estimators": [100,300, 500, 700, 900],
          "max_leaves" : [5,7,9], 
          'booster' : ['gblinear','DART','gbtree'],
 "max_depth" : [ 11, 12, 15],
 "gamma": [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "subsample":[0.3,0.5,0.7],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]}
xg = XGBRegressor()
rsearch = RandomizedSearchCV(xg,
                             param_distributions=params,n_jobs=-1,cv=5,verbose=3, n_iter=500, scoring='r2')
#xg = XGBRegressor(n_estimators = 2000, max_depth = 5, max_leaves=5, grow_policy='lossguide', booster = 'gblinear',
#                   random_state = 0, eval_metric = r2_score)
serch = rsearch.fit(x_train,y_train)
#model = xg.fit(x_train,y_train,eval_set=[(x_val, y_val)])
```

```python
from xgboost import XGBRegressor
model = XGBRegressor( booster='gbtree', colsample_bylevel=1.0,
       colsample_bynode=1.0, tree_method='auto',colsample_bytree=0.4, device='gpu',
                      eval_metric=r2_score, gamma=0.1, learning_rate=0.1, max_depth=12, max_leaves=9, n_estimators=900)
model.fit(x_train,y_train)
```
